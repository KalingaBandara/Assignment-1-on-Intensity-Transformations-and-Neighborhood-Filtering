```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv.imread('../Assignment1_Images/BrainProtonDensitySlice9.png', cv.IMREAD_GRAYSCALE)  


t1 = np.linspace(0,80,80)
t2 = np.linspace(81,150,0)
t3 = np.linspace(151,255,100)
t4 = np.linspace(180,255,76)
t = np.concatenate((t1,t2,t3,t4), axis=0).astype(np.uint8)

# Create the output image
output_image = cv.LUT(input_image, t)

# Plot the intensity transformation graph
plt.plot(range(256), t, label='Intensity Transformation')
plt.xlabel('Input Intensity') 
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')
plt.show()
# Save the plot as an image
plt.savefig('intensity_transformation_graph_Q2_a.png')


# Save or display the output image
cv.imwrite('Q2_output_white_matter.jpg', output_image)  # Save the output image
cv.imshow('Output Image', output_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
